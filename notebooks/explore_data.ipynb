```python
import pandas as pd

# Load the dataset
df_original = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Create a copy of the dataset for processing
df = df_original.copy()

print(df.head())

print(df.info())

print(df.isnull().sum()) # Only checks the NaN value

print(df.describe())

```

```python
# Check for empty strings in TotalCharges (cleaned dataset)
empty_strings = df['TotalCharges'].eq(' ')
print(f"\nNumber of empty strings in TotalCharges: {empty_strings.sum()}")
```

```python
# Convert TotalCharges to numeric (empty strings will become NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill NaN values with 0
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Check if TotalCharges has been converted to numeric and filled with 0
print(df['TotalCharges'].dtype)
empty_strings = df['TotalCharges'].eq(0)
print(f"\nNumber of empty strings in TotalCharges: {empty_strings.sum()}")
```
